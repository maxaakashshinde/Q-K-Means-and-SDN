```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# import math lib
from math import pi

# import Qiskit
from qiskit import Aer, IBMQ, execute
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister

# import basic plot tools
from qiskit.tools.visualization import plot_histogram
backend = Aer.get_backend('qasm_simulator')
```

```python
def get_theta(d):
    x = d[0]
    y = d[1]
    theta = math.acos((x+y)/2)
    return theta
```

```python
def get_Distance(x,y):
    theta_1 = get_theta(x)
    theta_2 = get_theta(y)
    qr = QuantumRegister(4, name="qr")
    
    cr = ClassicalRegister(1, name="cr")

    qc = QuantumCircuit(qr, cr, name="k_means")
    
    qc.h(qr[0])
    qc.h(qr[1])
    qc.h(qr[2])
    qc.u3(theta_1*pi, pi, 0, qr[1])
    qc.u3(theta_2*pi, pi, 0, qr[2])
    qc.cswap(qr[0], qr[1], qr[2])
    qc.h(qr[0])

    qc.measure(qr[0], cr[0])
    qc.draw()
    qc.reset(qr)

    job = execute(qc,backend=backend, shots=1024)
  
    result = job.result()
   
    data = result.data()['counts']
   
    if len(data)==1:
        return 0.0
    else:
        return data['0x0']/1024.0
    
```

```python
def get_data():
    points = datapoints
    centers = datapoints_label
    
    return points,centers
```

```python

import matplotlib.pyplot as plt
def draw_plot(points,centers,label=True):
    if label==False:
        plt.scatter(points[:,0], points[:,1])
    else:
        plt.scatter(points[:,0], points[:,1], c=centers, cmap='viridis')
    plt.xlim(0.37,0.385)
    plt.ylim(0.37,0.385)
    plt.show()
```

```python
def plot_centroids(centers):
    plt.scatter(centers[:,0], centers[:,1])
    plt.xlim(0.37,0.385)
    plt.ylim(0.37,0.385)
    plt.show()
```

```python
def initialize_centers(points,k):
    return points[np.random.randint(points.shape[0],size=k),:]
```

```python
def find_nearest_neighbour(points,centroids):
    
    n = len(points)
    k = centroids.shape[0]
    centers = np.zeros(n)
    
    for i in range(n):
        min_dis = 10000
        ind = 0
        for j in range(k):
            temp_dis = get_Distance(points[i,:],centroids[j,:])
            
            if temp_dis < min_dis:
                min_dis = temp_dis
                ind = j
        centers[i] = ind
    
    return centers
```

```python
def find_centroids(points,centers):
    n = len(points)
    k = int(np.max(centers))+1
   
    centroids = np.zeros([k,2])
    
    for i in range(k):
        centroids[i,:] = np.average(points[centers==i])
        
    return centroids
```

```python
def preprocess(points):
    n = len(points)
    x = 30.0*np.sqrt(2)
    for i in range(n):
        points[i,:]+=15
        points[i,:]/=x
    
    return points
```

```python
n = 5000  
k = 2     
std = 2   
points,o_centers = get_data()      
points = preprocess(points)        
fig = plt.figure()
fig.subplots_adjust(bottom=0.0000001)
draw_plot(points,o_centers,label=True)

centroids = initialize_centers(points,k)  

for i in range(10):
    
    centers = find_nearest_neighbour(points,centroids)       
    fig = plt.figure()
    fig.subplots_adjust(bottom=0.0000001)
    draw_plot(points,centers)
    centroids = find_centroids(points,centers)            
    plot_centroids(centroids)
    
```
